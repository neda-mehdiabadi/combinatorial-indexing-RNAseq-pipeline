```python
from datasets import Dataset
from sklearn.model_selection import train_test_split
import pandas as pd

file_path = '/data/mcri_heartv2/mcri_nedar/DCM-Foundation-Model-data/train_datasets/train_dataset/20250109_train_107samples_human_dcm_control_pediatric_adult_CMonly_4096_w_length.dataset/'

tokenized_dataset = Dataset.load_from_disk(file_path)


train_df = tokenized_dataset.to_pandas()

# Add a stratification group column based on `disease`, 'group' and `sex` (or any other features)
train_df['stratify_group'] = train_df['disease'] + "_" + train_df['group']+ "_" + train_df['sex']

# Stratified sampling: Select 10% of the dataset for hyperparameter optimization (for example)
train_subset, _ = train_test_split(
    train_df, 
    test_size=0.9,  # Keep 10% of data for hyperparameter optimization
    stratify=train_df['stratify_group'],  # Ensure stratification
    random_state=42
)

# Convert back to a Hugging Face Dataset object
train_subset_dataset = Dataset.from_pandas(train_subset)

# Now, you can use `train_subset_dataset` for hyperparameter optimization
train_subset_dataset.save_to_disk('/data/mcri_heartv2/mcri_nedar/DCM-Foundation-Model-data/Stratified Sampling/train_subset_dataset.dataset')

```

```python
from datasets import Dataset

# Load your dataset
file_path = '/data/mcri_heartv2/mcri_nedar/DCM-Foundation-Model-data/siRNA/datasets/20250227_invitro_RABGAP1L_multiClass_4096_w_length_edit.dataset/'
tokenized_dataset = Dataset.load_from_disk(file_path)

# Check the dataset to understand its structure
print(tokenized_dataset)

# Define a function to add the 'disease' column based on the 'condition' column
def add_disease(tk_data):
    # Check the value in the 'condition' column and assign the appropriate 'disease' value
    if tk_data['condition'] == 'patient_scrambled':
        tk_data['disease'] = 'DCM'  # Assign the value of 'Control' as disease for this condition
    else:
        tk_data['disease'] = 'Control'  # Assign the value of 'DCM' for other conditions
    return tk_data

# Apply the function to the dataset, creating the 'disease' column
tokenized_dataset = tokenized_dataset.map(add_disease)

# Check the updated dataset with the 'disease' column
print(tokenized_dataset[1])


#tokenized_dataset.save_to_disk('/data/mcri_heartv2/mcri_nedar/DCM-Foundation-Model-data/siRNA/datasets/20250227_invitro_RABGAP1L_multiClass_4096_w_length_edit.dataset/')
```
